Parar es cuando está parada y con la línea del cuello paralelo o en diagonal hacia arriba del piso. Cuatro patas en el piso o tres y una apoyada en la otra y no se está moviendo

Comer abajo es cuando está comiendo con la línea del cuello en diagonal hacia abajo y puede ser que haga uno o dos pasos y vuelva a comer

Comer arriba tiene la línea del cuello paralelo al piso o en diagonal hacia arriba

Comer abajo siempre tiene la línea del cuello en diagonal hacia abajo
